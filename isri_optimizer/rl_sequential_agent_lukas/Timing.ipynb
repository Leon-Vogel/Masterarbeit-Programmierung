```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("GNN_Timing_Data.csv", sep=";", index_col=0)
```

```python
data
```

```python
data['time_per_obs'] = data.proc_time / data.batch_size
```

```python
for column in data.columns.drop('time_per_obs'):
    # Create scatter plot
    plt.figure(figsize=(8, 6))
    plt.scatter(data[column], data['time_per_obs'], alpha=0.5)
    
    # Set plot title and labels
    plt.title(f'{column} vs proc_time')
    plt.xlabel(column)
    plt.ylabel('time_per_obs')
    
    # Show plot
    plt.grid(True)
    plt.show()
```
