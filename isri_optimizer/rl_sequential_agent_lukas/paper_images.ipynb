```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
```

# Results with 20 Jobs (like Training Data)

```python
rl_results = pd.read_csv("Test_Results_RL_20.csv", sep=";", index_col=0)
rl_results['percent_diff_diffsum'] = ((rl_results.rl_diffsum- rl_results.ga_diffsum) / rl_results.ga_diffsum) * 100
rl_results['percent_diff_tardiness'] = ((rl_results.rl_tardiness- rl_results.ga_tardiness) / rl_results.ga_tardiness) * 100
```

```python
rl_results
```

```python
print("Max Diffsum Diff Perc: ", rl_results.percent_diff_diffsum.max())
print("Min Diffsum Diff Perc: ", rl_results.percent_diff_diffsum.min())
print("Max Tardiness Diff Perc: ", rl_results.percent_diff_tardiness.max())
print("Min Tardiness Diff Perc: ", rl_results.percent_diff_tardiness.min())
```

```python
plt.style.use(['science', 'no-latex'])
```

```python
fig, axs = plt.subplots(1, 2, figsize=[8, 4])
fontname = 'Arial'
fontsize = 12
axs[0].hist(rl_results['percent_diff_diffsum'], bins=np.arange(-13.5, 1, 0.5), width=0.4, facecolor='dimgray')
fig.suptitle("Objective Difference GA vs. RL")
axs[0].set_xlabel("Workload Objective Gap (%)")
axs[0].set_ylabel("Number of Test-Instances")
axs[0].set_yticks(np.arange(0, 7, 2))
axs[0].set_yticklabels(np.arange(0, 7, 2))
axs[1].hist(rl_results['percent_diff_tardiness'], bins=np.arange(-7.5, 3.5, 0.5), width=0.4, facecolor='dimgray')
# axs[1].set_title("Distribution of the Deadline Gap")
axs[1].set_xlabel("Deadline Objective Gap (%)")
axs[1].set_ylim(0, 14)
# axs[1].set_ylabel("Number of Test-Instances")
plt.savefig('./images/objective_difference_20.svg', format='svg')
```

```python
plt.figure()
plt.boxplot(rl_results[rl_results.time_in_seconds < 1]['time_in_seconds'])
plt.title("RL Solving Time")
plt.xlabel("Time in Seconds")
plt.ylabel("Number of Test-Instances")
plt.show()
```

```python
plt.figure(figsize=[5, 3])
instance = 5
ga_run = pd.read_csv(f"savefiles/ga_stats/ga_{instance}_20_jobs.csv", sep=",", header=None)
ga_run.columns = ["diffsum_mean", "tardiness_mean", "diffsum_min", "tardiness_min", "diffsum_std", "tardiness_std",  "diffsum_selected", "tardiness_selected"]
rl_value = rl_results[rl_results.instance_id == instance].rl_diffsum.values[0]
plt.hlines(rl_value, 0, 200, color='green', linestyle='--', label="RL Result")
plt.plot(ga_run.diffsum_mean, linestyle='-', label='Workload Objective Mean')
plt.plot(ga_run.diffsum_selected, linestyle='-', label="Workload Objective Selected", color='red')
plt.fill_between(np.arange(0, len(ga_run), 1), ga_run.diffsum_mean + ga_run.diffsum_std,
                 ga_run.diffsum_mean - ga_run.diffsum_std, alpha=0.5,
                 color='grey', linestyle="--", label="Standard Deviation")
plt.legend()
plt.title("GA Run Workload Objective")
plt.xlabel("GA Generation")
plt.ylabel("Workload Objective Value")
plt.savefig('./images/GA_single_run_workload_20.svg', format='svg')
```

```python
plt.figure(figsize=[5, 3])
instance = 7
ga_run = pd.read_csv(f"./savefiles/ga_stats/ga_{instance}_20_jobs.csv", sep=",", header=None)
ga_run.columns = ["diffsum_mean", "tardiness_mean", "diffsum_min", "tardiness_min", "diffsum_std", "tardiness_std",  "diffsum_selected", "tardiness_selected"]
rl_value = rl_results[rl_results.instance_id == instance].rl_tardiness.values[0]
plt.hlines(rl_value, 0, 200, color='green', linestyle='--', label="RL Result")
plt.plot(ga_run.tardiness_mean, linestyle='-', label='Tardiness Objective Mean')
plt.plot(ga_run.tardiness_selected, linestyle='-', label="Tardiness Objective Selected", color='red')
plt.fill_between(np.arange(0, len(ga_run), 1), ga_run.tardiness_mean + ga_run.tardiness_std,
                 ga_run.tardiness_mean - ga_run.tardiness_std, alpha=0.5,
                 color='grey', linestyle="--", label="Standard Deviation")
plt.legend()
plt.title("GA Run Tardiness Objective")
plt.xlabel("GA Generation")
plt.ylabel("Tardiness Objective Value")
plt.savefig('./images/GA_single_run_tardiness_20.svg', format='svg')
```

```python
workload_equal_times = []
tardiness_equal_times = []
for test_id in rl_results.instance_id.to_list():
    ga_run = pd.read_csv(f"savefiles/ga_stats/ga_{test_id}_20_jobs.csv", sep=",", header=None)
    ga_run.columns = ["diffsum_mean", "tardiness_mean", "diffsum_min", "tardiness_min", "diffsum_std", "tardiness_std", "diffsum_selected", "tardiness_selected"]
    rl_value_diffsum = rl_results[rl_results.instance_id == instance].rl_diffsum.values[0]
    rl_value_tardiness = rl_results[rl_results.instance_id == instance].rl_tardiness.values[0]
    try:
        crosspoint_workload = np.where(ga_run.diffsum_selected < rl_value_diffsum)[0][0]
    except:
        crosspoint_workload = len(ga_run) + 1
    try:
        crosspoint_tardiness = np.where(ga_run.tardiness_selected < rl_value_tardiness)[0][0]
    except:
        crosspoint_tardiness = len(ga_run) + 1
    workload_equal_times.append(crosspoint_workload)
    tardiness_equal_times.append(crosspoint_tardiness)
    
```

```python
fig, axs = plt.subplots(1, 2, figsize=[5, 2])
fontname = 'Arial'
fontsize = 12
axs[0].hist(workload_equal_times, bins=np.arange(0, 211, 10), width=9.9, facecolor='dimgray')
fig.suptitle("Crosspoints")
axs[0].set_xlabel("Crosspoint Workload")
axs[0].set_ylabel("Number of Test-Instances")
axs[0].set_yticks(np.arange(0, 40, 5))
axs[0].set_yticklabels(np.arange(0, 40, 5))
axs[1].hist(tardiness_equal_times, bins=np.arange(0, 211, 10), width=9.9, facecolor='dimgray')
# axs[1].set_title("Distribution of the Deadline Gap")
axs[1].set_yticks(np.arange(0, 50, 5))
axs[1].set_yticklabels(np.arange(0, 50, 5))
axs[1].set_xlabel("Crosspoint Tardiness")
axs[1].set_ylim(0, 50)
# axs[1].set_ylabel("Number of Test-Instances")
plt.savefig('./images/crosspoints_20.svg', format='svg')
```
